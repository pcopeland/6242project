## Categorize Time of Day

```python
df = df.copy()
```

```python
import pandas as pd
def categorize_time(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    else:
        return 'Night'

# Ensure the column is in datetime format
df['FormattedStartTime_t'] = pd.to_datetime(df['FormattedStartTime_t'])

# Use a lambda function to apply 'categorize_time' on the hour extracted from 'FormattedStartTime_t'
df['TOD'] = df['FormattedStartTime_t'].dt.hour.apply(lambda x: categorize_time(x))
```

## Map Months to Seasons

```python
# define a function to map month to season
def categorize_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# Create 'Season' column
df['Season'] = df['Month'].apply(categorize_season)
```
